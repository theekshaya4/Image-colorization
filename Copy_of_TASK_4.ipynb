```python
from torchvision import transforms
from PIL import Image
import numpy as np

# Create a dummy image if test.jpg doesn't exist
try:
    img = Image.open("test.jpg")
except FileNotFoundError:
    print("test.jpg not found. Creating a placeholder image for demonstration.")
    data = np.random.randint(0, 255, (100, 100, 3), dtype=np.uint8)
    img = Image.fromarray(data)

aug = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(30),
])

for i in range(5):
    aug(img).save(f"aug_{i}.jpg")
print("Augmented images saved.")
```
